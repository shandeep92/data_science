```python
import numpy as np
import matplotlib.pyplot as plt

# Set the figure and axis colors to white
fig = plt.figure(facecolor='black')
ax = fig.add_subplot(111)
ax.set_facecolor('black')
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')
plt.rcParams['text.color'] = 'white'

# Generate two normal distributions with different means and standard deviations
men_aggression = np.random.normal(loc=0.5, scale=0.1, size=10000)
women_aggression = np.random.normal(loc=0.4, scale=0.1, size=10000)

# Calculate the 99th percentile aggression level for each distribution
men_99th_percentile = np.percentile(men_aggression, 99)
women_99th_percentile = np.percentile(women_aggression, 99)

# Plot the distributions and 99th percentile lines
plt.hist(men_aggression, bins=50, alpha=0.5, label='Men')
plt.hist(women_aggression, bins=50, alpha=0.5, label='Women')
plt.axvline(x=men_99th_percentile, color='white', linestyle='--', label='Men 99th percentile')
plt.axvline(x=women_99th_percentile, color='white', linestyle='--', label='Women 99th percentile')
plt.legend()
plt.xlabel('Aggression', color='white')
plt.ylabel('Frequency', color='white')
plt.show()


```
